```python
#Organisation Search    
@org_search.route('/search_organization',methods=['POST'])
def search_organization():
	data_status = {"responseStatus": 0, "result": ""}
	userId =request.json["userId"]
	orgName = request.json['orgName']
	if userId and orgName and request.method=="POST":
		try:
			user = PsgUsers.objects.get(id=userId)
			organizations = PiesaGivingOrganisations.objects(displayName__icontains=orgName,status__in=[1])
			queryset_users = PsgUsers.objects(Q(id__ne=userId)&Q(status__in=[1])&(Q(firstName__icontains=orgName) | Q(
                    lastName__icontains=orgName) | Q(firstName__icontains=orgName)))
			org_list = []
			users_list = []
			if organizations:
				for org in organizations:
					org_dict = {
					"id":str(org.id),
					"organizationName":org.displayName,
					"streetAddress":org.streetAddress,
					"cityId":str(org.cityId.id),
					"cityName":org.cityId.name,
					"profilePic":"http://3.133.43.47/media/images/organizations/proFile/"+str(org.profilePic)
					}
					org_list.append(org_dict)
			if queryset_users:
				for user in queryset_users:
					user_dict = {
					"id":str(user.id),
					"userName":user.userName,
					"firstName":user.firstName,
					"lastName":user.lastName,
					"privacy":user.privacy,
					}
					state = PiesaGivingStates.objects.get(stateId=user.state)
					profilePic = user.profilePic
					if profilePic != None:
						user_dict["profilePic"] = "http://3.133.43.47/media/images/user/" + str(
                            user.profilePic)
					else:
						user_dict["profilePic"] = ""
					user_dict["stateName"]=state.name
					users_list.append(user_dict)
			if org_list or users_list:
				data_status["responseStatus"]=1
				data_status["result"]="Organizations and Users fetched successfully!!"
				data_status["orgList"]=org_list
				data_status["userList"]=users_list
				return data_status
			else:
				data_status["responseStatus"]=0
				data_status["result"]="Organizations and Users not available!!!!"
				data_status["orgList"]=org_list
				data_status["userList"]=users_list
				return data_status
		except PsgUsers.DoesNotExist as e:
			data_status["responseStatus"]=0
			data_status["result"]="User Does not Exist!!!"
			data_status["orgList"]=org_list
			data_status["userList"]=users_list
			return data_status
		except Exception as e:
			data_status["responseStatus"]=0
			data_status["result"]=e.args[0]
			data_status["orgList"]=org_list
			data_status["userList"]=users_list
			return data_status
	else:
		data_status["responseStatus"]=0
		data_status["result"]="Required fields Missig!!"
		return data_status
```
